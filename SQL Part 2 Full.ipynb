```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('Baseball database.sqlite')
```

### 1.	Siapakah nama dari pemain yang memiliki gaji tertinggi?

```python
d = pd.read_sql_query("""SELECT P.name_first ||' '|| P.name_last AS Name, MAX(S.salary) AS Salary 
               FROM player P JOIN salary S ON S.player_id=P.player_id""",conn)
d
```

### 2.	Tom Kelly adalah salah satu pelatih yang mendapatkan penghargaan BBWAA Manager of the year. Siapa sajakah pemain yang bermain di tim yang dimanage oleh Kelly Ketika ia mendapatkan penghargaan tersebut?

```python
k = pd.read_sql_query("""SELECT DISTINCT name_first||' '||name_last as Name 
                FROM player P, fielding S,
                (SELECT team_id, year FROM manager 
                WHERE player_id=(SELECT player_id FROM player WHERE name_first='Tom' AND name_last='Kelly') AND 
                year =(SELECT year 
                FROM manager_award 
                WHERE player_id=(SELECT player_id FROM player WHERE name_first='Tom' AND name_last='Kelly') and award_id='BBWAA Manager of the year')
                )T
                WHERE S.player_id =P.player_id and S.team_id = T.team_id and S.year =T.year""",conn)
k
```

```python
c = pd.read_sql_query("""SELECT name_first||' '||name_last as Name 
                FROM player P, salary S,
                (SELECT team_id, year FROM manager 
                WHERE player_id=(SELECT player_id FROM player WHERE name_first='Tom' AND name_last='Kelly') AND 
                year =(SELECT year 
                FROM manager_award 
                WHERE player_id=(SELECT player_id FROM player WHERE name_first='Tom' AND name_last='Kelly') and award_id='BBWAA Manager of the year')
                )T
                WHERE S.player_id =P.player_id and S.team_id = T.team_id and S.year =T.year""",conn)
c
```

### 3.	Siapa saja pemain yang pernah ada dalam tim yang sama dengan Paul Abbot?

```python
a = pd.read_sql_query("""SELECT B.year, B.team_id, P.name_first||' '||P.name_last as Name 
                FROM player P, batting B,
                (SELECT team_id, year FROM batting 
                WHERE player_id=(SELECT player_id FROM player WHERE name_first='Paul' AND name_last='Abbott'))T 
                WHERE B.player_id=P.player_id AND T.year=B.year AND T.team_id=B.team_id""",conn)
a
```

### 4.	Siapa saja para pemain dari tim Brooklyn Atlantics yang masuk ke dalam hall of fame?

```python
n = pd.read_sql_query("""SELECT name_first||' '||name_last AS Name FROM
                        player P, hall_of_fame H,
                        (SELECT DISTINCT player_id FROM fielding M,
                        (SELECT DISTINCT team_id FROM team WHERE name='Brooklyn Atlantics')T
                        WHERE T.team_id = M.team_id)I
                        WHERE I.player_id = H.player_id AND H.player_id = P.player_id
                        """,conn)
n
```
